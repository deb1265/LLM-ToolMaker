```python
from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize

# Function to capitalize the first letter of a sentence
def capitalize_sentence(sentence):
    return sentence[0].capitalize() + sentence[1:] if sentence else sentence



```

```python
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5")

# Example input paragraph
paragraphs = [
    "As energy costs continue to rise and concerns about the environment grow, more and more homeowners on Long Island are turning to solar power as a sustainable and cost-effective solution. One of the key factors driving this trend is the availability of net metering through PSEG Long Island, which allows solar panel owners to offset their electricity bills by sending excess energy back to the grid. In this article, we’ll explore how Patriot Energy Solutions, a leading solar installation company, is helping homeowners leverage PSEG’s net metering policies to maximize their savings and reduce their carbon footprint."
]

for paragraph in paragraphs:
    print("=" * 100)
    print("Input_paragraph: ", paragraph)
    print("=" * 100)

    # Split the paragraph into individual sentences
    sentences = sent_tokenize(paragraph)

    paraphrased_sentences = []

    for sentence in sentences:
        print("-" * 100)
        print("Input_sentence: ", sentence)
        print("-" * 100)
        try:
            # Generate paraphrases
            para_phrases = parrot.augment(
                input_phrase=sentence,
                use_gpu=True,
                do_diverse=True,             # Enable this to get more diverse paraphrases
                adequacy_threshold=0.80,     # Lower this number if no paraphrases returned
                fluency_threshold=0.80
            )

            # Extract paraphrases and their scores
            if para_phrases:
                # Assuming the format (paraphrase, score)
                best_paraphrase = max(para_phrases, key=lambda x: x[1])
                paraphrased_sentences.append(capitalize_sentence(best_paraphrase[0]))
                print("Best Paraphrase:", best_paraphrase[0])
                print("Diversity Score:", best_paraphrase[1])
            else:
                print("No paraphrases returned")
                paraphrased_sentences.append(capitalize_sentence(sentence))  # Append original sentence if no paraphrases are returned
        except Exception as e:
            print("Error:", str(e))
            paraphrased_sentences.append(capitalize_sentence(sentence))  # Append original sentence if an error occurs

    # Join the paraphrased sentences back into a single paragraph
    paraphrased_paragraph = '. '.join(paraphrased_sentences) + '.'
    print("\nParaphrased_paragraph: ", paraphrased_paragraph)
```
